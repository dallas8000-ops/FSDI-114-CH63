```python
def count_even_numbers(collection):
    """Returns the count of numbers in the collection that are divisible by 2 (modulus 0)."""
    return sum(1 for n in collection if n % 2 == 0)

numbers = [3,5,27,20,6,3.5,60]
total = count_even_numbers(numbers)
print(f"total={total}")
```
